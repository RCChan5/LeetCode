**1071. Greatest Common Divisor of Strings**




```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        if str1 + str2 != str2 + str1:
            return ""
        from math import gcd
        return str1[:gcd(len(str1), len(str2))]
```

The key to solving this problem lies in applying mathematical principles, specifically the concept of the GCD. By considering the lengths of each string and utilizing the GCD of those lengths, we can effectively determine how to segment the strings.
