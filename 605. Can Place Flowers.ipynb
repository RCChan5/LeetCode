Problem Overview:

The goal of the function is to determine if its possible to plant at least n flowers in a flowerbed, following these rules:

A flower can only be planted in a 0 spot.
No two flowers can be planted in adjacent positions (i.e., no two consecutive 1s).


Why This Code Works:

The code correctly checks if each spot is available to plant a flower by ensuring both neighbors (if any) are 0.

It updates the flowerbed in-place to reflect the newly planted flowers.
